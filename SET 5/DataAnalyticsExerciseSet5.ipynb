```python
tab = et.parse("simple.xml")
root = tab.getroot()

# columns names
df_col =["id", "brand", "model", "year", "owned_by", "acquired", "price"]
rows = []

# iterate through each node and extract data
for node in root:
    data = {}
    for element in node:
        if element.tag in df_col:
            data[element.tag] = element.text
    rows.append(data)

# create DF
new_df = pd.DataFrame(rows, columns=df_col)
new_df
```
